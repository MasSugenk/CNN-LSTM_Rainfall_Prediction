```python
# ============================
# MOdel Pembanding CNN-LSTM
# ============================

import os
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, roc_auc_score
)
from sklearn.utils.class_weight import compute_class_weight

from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# ===============================
# 0. Reproducibility
# ===============================
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ===============================
# 1. Paths & Config
# ===============================
DATA_PATH = r'./Data_Sensor/sensor_curah_corrected_finalV2.csv'
OUT_DIR = r'./model_pembanding_CNN-LSTM_1jam'
os.makedirs(OUT_DIR, exist_ok=True)

BEST_MODEL_PATH = os.path.join(OUT_DIR, 'best_model.keras')
SCALER_PATH = os.path.join(OUT_DIR, 'scaler_X.joblib')
HISTORY_CSV = os.path.join(OUT_DIR, 'training_history.csv')
TEST_PRED_CSV = os.path.join(OUT_DIR, 'test_predictions.csv')
BEST_THR_TXT = os.path.join(OUT_DIR, 'best_threshold.txt')

# ===============================
# 2. Parameters
# ===============================
time_steps = 12        # 1 jam history (12 x 5 menit)
horizon_steps = 12     # 1 jam ke depan
test_ratio = 0.2
batch_size = 64
epochs = 500

# ===============================
# 3. Load & Prepare Data
# ===============================
df = pd.read_csv(DATA_PATH)
df.columns = [c.strip() for c in df.columns]

# Build Datetime
if 'Datetime' not in df.columns:
    df['Datetime'] = pd.to_datetime(
        df['Tanggal'].astype(str) + ' ' + df['Jam'].astype(str),
        dayfirst=True,
        errors='coerce'
    )

df = df.sort_values('Datetime').reset_index(drop=True)

# ===============================
# 4. Target & Features (FOCUS)
# ===============================
TARGET_COL = 'Curah_Hujan_corrected'

FEATURES = [
    'Curah_Hujan_corrected',
    'Curah_Hujan_original',
    'Suhu',
    'Kelembaban',
    'Kecepatan_Angin'
]

# Validasi kolom
missing = [c for c in FEATURES + [TARGET_COL] if c not in df.columns]
if missing:
    raise ValueError(f"Kolom tidak ditemukan: {missing}")

# Konversi numeric
for c in FEATURES + [TARGET_COL]:
    df[c] = pd.to_numeric(df[c], errors='coerce')

df.interpolate(method='linear', inplace=True, limit_direction='both')
df.dropna(subset=FEATURES + [TARGET_COL], inplace=True)
df.reset_index(drop=True, inplace=True)

# Binary label
df['Hujan'] = (df[TARGET_COL] > 0).astype(int)
print("Distribusi kelas:", df['Hujan'].value_counts().to_dict())

# ===============================
# 5. Raw X & y (BELUM DISCALING)
# ===============================
X_raw = df[FEATURES].values
y_raw = df['Hujan'].values

# ===============================
# 6. Create Sequences
# ===============================
def create_sequences(X, y, time_steps, horizon_steps):
    Xs, ys, idxs = [], [], []
    max_i = len(X) - time_steps - horizon_steps + 1

    for i in range(max_i):
        Xs.append(X[i:i+time_steps])
        label_idx = i + time_steps + horizon_steps - 1
        ys.append(y[label_idx])
        idxs.append(label_idx)

    return np.array(Xs), np.array(ys), np.array(idxs)

X_seq, y_seq, idx_seq = create_sequences(
    X_raw, y_raw,
    time_steps, horizon_steps
)

print("Sequence shape:", X_seq.shape)

# ===============================
# 7. TIME-BASED SPLIT
# ===============================
split_idx = int(len(X_seq) * (1 - test_ratio))

X_train = X_seq[:split_idx]
X_test  = X_seq[split_idx:]

y_train = y_seq[:split_idx]
y_test  = y_seq[split_idx:]

idx_test = idx_seq[split_idx:]

print("Train samples:", len(y_train))
print("Test samples :", len(y_test))

# ===============================
# 8. Scaling (FIT ONLY ON TRAIN)
# ===============================
scaler = StandardScaler()

X_train_2d = X_train.reshape(-1, X_train.shape[2])
X_test_2d  = X_test.reshape(-1, X_test.shape[2])

scaler.fit(X_train_2d)

X_train = scaler.transform(X_train_2d).reshape(X_train.shape)
X_test  = scaler.transform(X_test_2d).reshape(X_test.shape)

joblib.dump(scaler, SCALER_PATH)

# ===============================
# 9. Class Weights
# ===============================
classes = np.unique(y_train)
cw = compute_class_weight('balanced', classes=classes, y=y_train)
class_weight = {int(c): float(w) for c, w in zip(classes, cw)}
print("Class weight:", class_weight)

# ===============================
# 10. Build CNN-LSTM Hybrid Model
# ===============================
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, LSTM, Dense, Dropout, MultiHeadAttention, LayerNormalization, GlobalAveragePooling1D
from tensorflow.keras.models import Model

def create_attention_cnn_lstm(time_steps, num_features):
    # 1. Input Layer (Sesuai dengan matriks sequence kamu)
    inputs = Input(shape=(time_steps, num_features))
    
    # 2. CNN Layers (Feature Extraction Spasial/Sensor)
    # Gunakan padding='same' agar dimensi runtun waktu tidak habis menyusut
    x = Conv1D(filters=64, kernel_size=3, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x) # Mengurangi noise, sisa 6 time steps
    
    x = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same')(x)
    
    # 3. LSTM Layer (Temporal Modeling)
    # PENTING: return_sequences=True WAJIB aktif agar output memiliki 
    # dimensi waktu yang bisa dibaca oleh Layer Attention
    lstm_out = LSTM(64, return_sequences=True)(x) 
    
    # 4. Mechanism Attention (Multi-Head Attention)
    # Model akan belajar 'step' waktu mana yang paling krusial memicu hujan
    attention_out = MultiHeadAttention(num_heads=4, key_dim=16)(lstm_out, lstm_out)
    
    # Residual Connection & Layer Normalization untuk stabilitas training
    x = LayerNormalization()(attention_out + lstm_out)
    
    # 5. Fully Connected Layers
    # Menggunakan GlobalAveragePooling1D lebih aman dari overfitting dibanding Flatten()
    x = GlobalAveragePooling1D()(x)
    x = Dense(32, activation='relu')(x)
    x = Dropout(0.3)(x)
    
    # 6. Output Layer (Klasifikasi biner: Hujan / Tidak Hujan)
    outputs = Dense(1, activation='sigmoid')(x)
    
    # Bangun Model
    model = Model(inputs=inputs, outputs=outputs)
    return model

# ===============================
# 11. Callbacks
# ===============================
callbacks = [
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', patience=7, factor=0.5, min_lr=1e-6),
    ModelCheckpoint(BEST_MODEL_PATH, monitor='val_loss', save_best_only=True)
]

# ===============================
# 12. Train Model
# ===============================
# Tentukan dimensi input berdasarkan data kamu
num_features = X_train.shape[2] # otomatis mendeteksi 5 fitur kamu

# Panggil fungsi model baru
model = create_attention_cnn_lstm(time_steps, num_features)

# Kompilasi tetap menggunakan pola lamamu yang sudah benar
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# Proses training (fit) tetap sama menggunakan class_weight kamu
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=epochs,
    batch_size=batch_size,
    class_weight=class_weight, # Pertahankan ini karena data kamu imbalanced
    callbacks=callbacks # Pertahankan EarlyStopping & ReduceLROnPlateau kamu
)

pd.DataFrame(history.history).to_csv(HISTORY_CSV, index=False)

# ===============================
# 13. Evaluation + Best Threshold
# ===============================
y_prob = model.predict(X_test).flatten()

thr_grid = np.arange(0.1, 0.9, 0.01)
best_thr, best_f1 = 0.5, -1

for thr in thr_grid:
    y_pred = (y_prob >= thr).astype(int)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    if f1 > best_f1:
        best_f1 = f1
        best_thr = thr

y_best = (y_prob >= best_thr).astype(int)

print("\n=== BEST THRESHOLD (F1) ===")
print("Threshold :", best_thr)
print("Accuracy :", accuracy_score(y_test, y_best))
print("Precision:", precision_score(y_test, y_best, zero_division=0))
print("Recall   :", recall_score(y_test, y_best, zero_division=0))
print("F1-score :", f1_score(y_test, y_best, zero_division=0))
print("ROC AUC  :", roc_auc_score(y_test, y_prob))
print("Confusion matrix:\n", confusion_matrix(y_test, y_best))

with open(BEST_THR_TXT, "w") as f:
    f.write(str(best_thr))

# ===============================
# 14. Save Test Predictions
# ===============================
df_out = df.iloc[idx_test].copy().reset_index(drop=True)
df_out['y_true'] = y_test
df_out['pred_prob'] = y_prob
df_out['pred_best'] = y_best

df_out.to_csv(TEST_PRED_CSV, index=False)

# ===============================
# 15. Plot Curves
# ===============================
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()

plt.tight_layout()
plt.show()

print("\nTRAINING SELESAI — MODEL TIME-BASED SIAP DIGUNAKAN")
print("Model  :", BEST_MODEL_PATH)
print("Scaler :", SCALER_PATH)
```

```python
# ==============================================================================
# JURNAL SINTA 1/2 EXPERIMENT: Attention-based CNN-LSTM untuk Prediksi Hujan
# ==============================================================================

import os
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, roc_auc_score
)
from sklearn.utils.class_weight import compute_class_weight

from tensorflow.keras.layers import (
    Input, Conv1D, MaxPooling1D, SpatialDropout1D, LSTM, 
    MultiHeadAttention, LayerNormalization, GlobalAveragePooling1D, Dense, Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# ==============================================================================
# 0. Reproducibility (Penting untuk Validitas Jurnal ilmiah)
# ==============================================================================
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ==============================================================================
# 1. Paths & Config
# ==============================================================================
DATA_PATH = r'./Data_Sensor/sensor_curah_corrected_finalV2.csv'
OUT_DIR = r'./model_pembanding_CNN-LSTM_1jam'
os.makedirs(OUT_DIR, exist_ok=True)

BEST_MODEL_PATH = os.path.join(OUT_DIR, 'best_model.keras')
SCALER_PATH = os.path.join(OUT_DIR, 'scaler_X.joblib')
HISTORY_CSV = os.path.join(OUT_DIR, 'training_history.csv')
TEST_PRED_CSV = os.path.join(OUT_DIR, 'test_predictions.csv')
BEST_THR_TXT = os.path.join(OUT_DIR, 'best_threshold.txt')

# ==============================================================================
# 2. Parameters
# ==============================================================================
time_steps = 12        # 1 jam history (12 x 5 menit)
horizon_steps = 12     # 1 jam ke depan
test_ratio = 0.2
batch_size = 64
epochs = 100           # EarlyStopping akan bertindak sebagai rem otomatis jika overfitting

# ==============================================================================
# 3. Load & Prepare Data
# ==============================================================================
df = pd.read_csv(DATA_PATH)
df.columns = [c.strip() for c in df.columns]

# Build Datetime
if 'Datetime' not in df.columns:
    df['Datetime'] = pd.to_datetime(
        df['Tanggal'].astype(str) + ' ' + df['Jam'].astype(str),
        dayfirst=True,
        errors='coerce'
    )

df = df.sort_values('Datetime').reset_index(drop=True)

# ==============================================================================
# 4. Target & Features
# ==============================================================================
TARGET_COL = 'Curah_Hujan_corrected'

FEATURES = [
    'Curah_Hujan_corrected',
    'Curah_Hujan_original',
    'Suhu',
    'Kelembaban',
    'Kecepatan_Angin'
]

# Validasi kolom
missing = [c for c in FEATURES + [TARGET_COL] if c not in df.columns]
if missing:
    raise ValueError(f"Kolom tidak ditemukan: {missing}")

# Konversi ke numeric & pastikan tipe data float eksplisit untuk menghindari FutureWarning Pandas
for c in FEATURES + [TARGET_COL]:
    df[c] = pd.to_numeric(df[c], errors='coerce')

df[FEATURES] = df[FEATURES].astype(float)

# Interpolasi aman dari warning
df.interpolate(method='linear', inplace=True, limit_direction='both')
df.dropna(subset=FEATURES + [TARGET_COL], inplace=True)
df.reset_index(drop=True, inplace=True)

# Binary label (Hujan / Tidak Hujan)
df['Hujan'] = (df[TARGET_COL] > 0).astype(int)
print("Distribusi kelas riil:", df['Hujan'].value_counts().to_dict())

# Extract raw arrays
X_raw = df[FEATURES].values
y_raw = df['Hujan'].values

# ==============================================================================
# 5. Create Sequences (Sliding Window)
# ==============================================================================
def create_sequences(X, y, time_steps, horizon_steps):
    Xs, ys, idxs = [], [], []
    max_i = len(X) - time_steps - horizon_steps + 1

    for i in range(max_i):
        Xs.append(X[i:i+time_steps])
        label_idx = i + time_steps + horizon_steps - 1
        ys.append(y[label_idx])
        idxs.append(label_idx)

    return np.array(Xs), np.array(ys), np.array(idxs)

X_seq, y_seq, idx_seq = create_sequences(X_raw, y_raw, time_steps, horizon_steps)
print("Sequence shape:", X_seq.shape)

# ==============================================================================
# 6. Time-Based Split (Validasi runtun waktu yang ketat)
# ==============================================================================
split_idx = int(len(X_seq) * (1 - test_ratio))

X_train, X_test  = X_seq[:split_idx], X_seq[split_idx:]
y_train, y_test  = y_seq[:split_idx], y_seq[split_idx:]
idx_test         = idx_seq[split_idx:]

print("Train samples:", len(y_train))
print("Test samples :", len(y_test))

# ==============================================================================
# 7. Scaling (Mencegah Data Leakage: Fit hanya pada Train)
# ==============================================================================
scaler = StandardScaler()

X_train_2d = X_train.reshape(-1, X_train.shape[2])
X_test_2d  = X_test.reshape(-1, X_test.shape[2])

scaler.fit(X_train_2d)

X_train = scaler.transform(X_train_2d).reshape(X_train.shape)
X_test  = scaler.transform(X_test_2d).reshape(X_test.shape)

joblib.dump(scaler, SCALER_PATH)

# ==============================================================================
# 8. Class Weights (Penanganan Imbalanced Data)
# ==============================================================================
classes = np.unique(y_train)
cw = compute_class_weight('balanced', classes=classes, y=y_train)
class_weight = {int(c): float(w) for c, w in zip(classes, cw)}
print("Class weight penyeimbang:", class_weight)

# ==============================================================================
# 9. Arsitektur Model: Attention-based CNN-LSTM (Model Usulan/Novelty)
# ==============================================================================
def create_attention_cnn_lstm(time_steps, num_features):
    inputs = Input(shape=(time_steps, num_features))
    
    # --- Feature Extraction Spasial (CNN) ---
    x = Conv1D(filters=64, kernel_size=3, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    x = SpatialDropout1D(0.2)(x)  # Mencegah overfitting korelasi sensorik
    
    x = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same')(x)
    x = SpatialDropout1D(0.2)(x)
    
    # --- Temporal Modeling (LSTM) ---
    lstm_out = LSTM(64, return_sequences=True)(x) 
    
    # --- Mechanism Attention (Multi-Head Attention) ---
    # Model fokus secara dinamis pada rentang waktu krusial pemicu hujan
    attention_out = MultiHeadAttention(num_heads=4, key_dim=16)(lstm_out, lstm_out)
    
    # Residual Connection & Layer Normalization untuk stabilitas gradien
    x = LayerNormalization()(attention_out + lstm_out)
    
    # --- Classification Layers ---
    x = GlobalAveragePooling1D()(x) # Lebih aman dari overfitting dibanding Flatten()
    x = Dense(32, activation='relu')(x)
    x = Dropout(0.4)(x)            # Ditingkatkan ke 0.4 untuk menekan overfitting di Dense layer
    
    outputs = Dense(1, activation='sigmoid')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

# ==============================================================================
# 10. Callbacks & Inisialisasi Model
# ==============================================================================
callbacks = [
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', patience=7, factor=0.5, min_lr=1e-6),
    ModelCheckpoint(BEST_MODEL_PATH, monitor='val_loss', save_best_only=True)
]

num_features = X_train.shape[2]
model = create_attention_cnn_lstm(time_steps, num_features)

# Menggunakan learning rate 0.0005 agar konvergensi model lebih stabil
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==============================================================================
# 11. Proses Training Model
# ==============================================================================
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=epochs,
    batch_size=batch_size,
    class_weight=class_weight,
    callbacks=callbacks,
    verbose=1
)

pd.DataFrame(history.history).to_csv(HISTORY_CSV, index=False)

# ==============================================================================
# 12. Evaluasi & Optimasi Threshold Berbasis F1-Score
# ==============================================================================
y_prob = model.predict(X_test).flatten()

thr_grid = np.arange(0.1, 0.9, 0.01)
best_thr, best_f1 = 0.5, -1

# Mencari koordinat threshold terbaik untuk mengatasi ketimpangan kelas minoritas (hujan)
for thr in thr_grid:
    y_pred = (y_prob >= thr).astype(int)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    if f1 > best_f1:
        best_f1 = f1
        best_thr = thr

y_best = (y_prob >= best_thr).astype(int)

print("\n" + "="*40)
print("=== HASIL EVALUASI OPTIMAL (ATTENTION-CNN-LSTM) ===")
print("="*40)
print(f"Best Threshold (F1) : {best_thr:.2f}")
print(f"Accuracy            : {accuracy_score(y_test, y_best):.4f}")
print(f"Precision           : {precision_score(y_test, y_best, zero_division=0):.4f}")
print(f"Recall              : {recall_score(y_test, y_best, zero_division=0):.4f}")
print(f"F1-score            : {f1_score(y_test, y_best, zero_division=0):.4f}")
print(f"ROC AUC             : {roc_auc_score(y_test, y_prob):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_best))
print("="*40)

with open(BEST_THR_TXT, "w") as f:
    f.write(str(best_thr))

# ==============================================================================
# 13. Save Hasil Prediksi Uji
# ==============================================================================
df_out = df.iloc[idx_test].copy().reset_index(drop=True)
df_out['y_true'] = y_test
df_out['pred_prob'] = y_prob
df_out['pred_best'] = y_best

df_out.to_csv(TEST_PRED_CSV, index=False)

# ==============================================================================
# 14. Visualisasi Kurva Pembelajaran (Wajib dicantumkan di Paper)
# ==============================================================================
plt.figure(figsize=(12, 5))

# Plot Akurasi
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Val Accuracy', color='orange', linestyle='--', linewidth=2)
plt.title('Kurva Akurasi Model', fontsize=12)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, linestyle=':')

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', color='blue', linewidth=2)
plt.plot(history.history['val_loss'], label='Val Loss', color='orange', linestyle='--', linewidth=2)
plt.title('Kurva Loss Model', fontsize=12)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, linestyle=':')

plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, 'Learning_Curves.png'), dpi=300)
plt.show()

print("\nTRAINING DAN EVALUASI SELESAI — GRAFIK BELAJAR BERHASIL DISIMPAN.")
```

```python
# ==============================================================================
# JURNAL SINTA 1/2 EXPERIMENT: HIGHLY OPTIMIZED ATTENTION-CNN-LSTM
# Skenario: Dilated CNN + Residual Multi-Head Attention + Time Feature Engineering
# ==============================================================================

import os
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, roc_auc_score
)
from sklearn.utils.class_weight import compute_class_weight

from tensorflow.keras.layers import (
    Input, Conv1D, SpatialDropout1D, LSTM, 
    MultiHeadAttention, LayerNormalization, GlobalAveragePooling1D, Dense, Dropout, Add
)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# ==============================================================================
# 0. Reproducibility (Penting untuk Validitas Ilmiah)
# ==============================================================================
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ==============================================================================
# 1. Paths & Config
# ==============================================================================
DATA_PATH = r'./Data_Sensor/sensor_curah_corrected_finalV2.csv'
OUT_DIR = r'./model_pembanding_CNN-LSTM_1jam_optimized'
os.makedirs(OUT_DIR, exist_ok=True)

BEST_MODEL_PATH = os.path.join(OUT_DIR, 'best_model_optimized.keras')
SCALER_PATH = os.path.join(OUT_DIR, 'scaler_X_optimized.joblib')
HISTORY_CSV = os.path.join(OUT_DIR, 'training_history_optimized.csv')
TEST_PRED_CSV = os.path.join(OUT_DIR, 'test_predictions_optimized.csv')
BEST_THR_TXT = os.path.join(OUT_DIR, 'best_threshold_optimized.txt')

# ==============================================================================
# 2. Parameters
# ==============================================================================
time_steps = 12        # 1 jam history (12 x 5 menit)
horizon_steps = 12     # 1 jam ke depan
test_ratio = 0.2
batch_size = 64
epochs = 500           

# ==============================================================================
# 3. Load & Prepare Data
# ==============================================================================
df = pd.read_csv(DATA_PATH)
df.columns = [c.strip() for c in df.columns]

# Build Datetime secara aman dan presisi
if 'Datetime' not in df.columns:
    df['Datetime'] = pd.to_datetime(
        df['Tanggal'].astype(str) + ' ' + df['Jam'].astype(str),
        dayfirst=True,
        errors='coerce'
    )
else:
    df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')

# Hapus baris jika Datetime gagal terbentuk sebelum sorting
df.dropna(subset=['Datetime'], inplace=True)
df = df.sort_values('Datetime').reset_index(drop=True)

# --- ADVANCED FEATURE ENGINEERING FOR TIME-SERIES (DIURNAL SIKLUS) ---
# Menggunakan dt.hour + dt.minute/60.0 agar mendapatkan representasi waktu yang kontinu
jam_kontinu = df['Datetime'].dt.hour + df['Datetime'].dt.minute / 60.0
df['Jam_sin'] = np.sin(2 * np.pi * jam_kontinu / 24.0)
df['Jam_cos'] = np.cos(2 * np.pi * jam_kontinu / 24.0)

# ==============================================================================
# 4. Target & Final Features Selection
# ==============================================================================
TARGET_COL = 'Curah_Hujan_corrected'

FEATURES = [
    'Curah_Hujan_corrected',
    'Curah_Hujan_original',
    'Suhu',
    'Kelembaban',
    'Kecepatan_Angin',
    'Jam_sin',   
    'Jam_cos'    
]

# Validasi ketersediaan kolom
missing = [c for c in FEATURES + [TARGET_COL] if c not in df.columns]
if missing:
    raise ValueError(f"Kolom tidak ditemukan di dataset: {missing}")

# Konversi semua ke numeric dan float secara eksplisit guna menghindari warning Pandas
for c in FEATURES + [TARGET_COL]:
    df[c] = pd.to_numeric(df[c], errors='coerce')
df[FEATURES] = df[FEATURES].astype(float)

# Pembersihan data dan interpolasi linier runtun waktu
df.interpolate(method='linear', inplace=True, limit_direction='both')
df.dropna(subset=FEATURES + [TARGET_COL], inplace=True)
df.reset_index(drop=True, inplace=True)

# Membuat target class binary (1 jika hujan, 0 jika tidak)
df['Hujan'] = (df[TARGET_COL] > 0).astype(int)
print("Distribusi kelas pasca optimasi:", df['Hujan'].value_counts().to_dict())

X_raw = df[FEATURES].values
y_raw = df['Hujan'].values

# ==============================================================================
# 5. Create Sequences (Sliding Window) dengan Validasi Dimensi
# ==============================================================================
def create_sequences(X, y, time_steps, horizon_steps):
    Xs, ys, idxs = [], [], []
    max_i = len(X) - time_steps - horizon_steps + 1

    if max_i <= 0:
        raise ValueError("Jumlah data terlalu sedikit untuk parameter time_steps dan horizon_steps yang diberikan!")

    for i in range(max_i):
        Xs.append(X[i:i+time_steps])
        label_idx = i + time_steps + horizon_steps - 1
        ys.append(y[label_idx])
        idxs.append(label_idx)

    return np.array(Xs, dtype=np.float32), np.array(ys, dtype=np.int32), np.array(idxs, dtype=np.int32)

X_seq, y_seq, idx_seq = create_sequences(X_raw, y_raw, time_steps, horizon_steps)
print("Sequence shape (Optimized Input):", X_seq.shape)

# Verifikasi protektif dimensi array sebelum proses selanjutnya
if len(X_seq.shape) != 3:
    raise ValueError(f"Kegagalan struktur! Diharapkan array 3D, tetapi mendapatkan shape {X_seq.shape}")

# ==============================================================================
# 6. Time-Based Split (Validasi Ketat Runtun Waktu)
# ==============================================================================
split_idx = int(len(X_seq) * (1 - test_ratio))

X_train, X_test  = X_seq[:split_idx], X_seq[split_idx:]
y_train, y_test  = y_seq[:split_idx], y_seq[split_idx:]
idx_test         = idx_seq[split_idx:]

print("Train samples:", len(y_train))
print("Test samples :", len(y_test))

# ==============================================================================
# 7. Scaling (Aman dari IndexError karena Dimensi Diproteksi)
# ==============================================================================
scaler = StandardScaler()

# Membaca informasi bentuk dimensi 3D asli
n_samples_train, n_timesteps, n_features = X_train.shape
n_samples_test, _, _ = X_test.shape

# Flatten menjadi 2D hanya untuk proses transformasi scikit-learn
X_train_2d = X_train.reshape(-1, n_features)
X_test_2d  = X_test.reshape(-1, n_features)

# Fit scaler eksklusif pada data training saja (Mencegah Data Leakage)
scaler.fit(X_train_2d)

# Kembalikan struktur array ke format 3D asli (samples, time_steps, features)
X_train = scaler.transform(X_train_2d).reshape(n_samples_train, n_timesteps, n_features)
X_test  = scaler.transform(X_test_2d).reshape(n_samples_test, n_timesteps, n_features)

joblib.dump(scaler, SCALER_PATH)
print("Proses Scaling berhasil diselesaikan tanpa kendala dimensi.")

# ==============================================================================
# 8. Class Weights (Penanganan Ketimpangan Data Cuaca)
# ==============================================================================
classes = np.unique(y_train)
cw = compute_class_weight('balanced', classes=classes, y=y_train)
class_weight = {int(c): float(w) for c, w in zip(classes, cw)}
print("Class weight:", class_weight)

# ==============================================================================
# 9. Arsitektur Model: OPTIMIZED ATTENTION-CNN-LSTM (Dilated & Residual Block)
# ==============================================================================
def create_optimized_attention_cnn_lstm(time_steps, num_features):
    inputs = Input(shape=(time_steps, num_features))
    
    # Feature Spasial via Dilated CNN (Tanpa MaxPooling untuk menjaga resolusi waktu)
    cnn1 = Conv1D(filters=64, kernel_size=3, activation='relu', padding='same')(inputs)
    cnn1 = SpatialDropout1D(0.2)(cnn1)
    
    cnn2 = Conv1D(filters=32, kernel_size=3, activation='relu', padding='same', dilation_rate=2)(cnn1)
    cnn2 = SpatialDropout1D(0.2)(cnn2)
    
    # Temporal Modeling (LSTM)
    lstm_out = LSTM(64, return_sequences=True)(cnn2) 
    
    # Advanced Residual Multi-Head Attention (8 Heads, key_dim=32)
    attention_out = MultiHeadAttention(num_heads=8, key_dim=32)(lstm_out, lstm_out)
    
    # Residual Blok Add + Layer Normalization
    x = Add()([attention_out, lstm_out])
    x = LayerNormalization()(x)
    
    # Classification Head
    x = GlobalAveragePooling1D()(x) 
    x = Dense(64, activation='relu')(x) 
    x = Dropout(0.4)(x)            
    
    outputs = Dense(1, activation='sigmoid')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

# ==============================================================================
# 10. Callbacks & Inisialisasi Model
# ==============================================================================
callbacks = [
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', patience=7, factor=0.5, min_lr=1e-6),
    ModelCheckpoint(BEST_MODEL_PATH, monitor='val_loss', save_best_only=True)
]

model = create_optimized_attention_cnn_lstm(n_timesteps, n_features)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==============================================================================
# 11. Proses Training Model
# ==============================================================================
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=epochs,
    batch_size=batch_size,
    class_weight=class_weight,
    callbacks=callbacks,
    verbose=1
)

pd.DataFrame(history.history).to_csv(HISTORY_CSV, index=False)

# ==============================================================================
# 12. Evaluasi & Optimasi Threshold Berbasis F1-Score
# ==============================================================================
y_prob = model.predict(X_test).flatten()

thr_grid = np.arange(0.1, 0.9, 0.01)
best_thr, best_f1 = 0.5, -1

for thr in thr_grid:
    y_pred = (y_prob >= thr).astype(int)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    if f1 > best_f1:
        best_f1 = f1
        best_thr = thr

y_best = (y_prob >= best_thr).astype(int)

print("\n" + "="*50)
print("=== HASIL EVALUASI OPTIMAL (REVISED MODEL) ===")
print("="*50)
print(f"Best Threshold (F1) : {best_thr:.2f}")
print(f"Accuracy            : {accuracy_score(y_test, y_best):.4f}")
print(f"Precision           : {precision_score(y_test, y_best, zero_division=0):.4f}")
print(f"Recall              : {recall_score(y_test, y_best, zero_division=0):.4f}")
print(f"F1-score            : {f1_score(y_test, y_best, zero_division=0):.4f}")
print(f"ROC AUC             : {roc_auc_score(y_test, y_prob):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_best))
print("="*50)

with open(BEST_THR_TXT, "w") as f:
    f.write(str(best_thr))

# ==============================================================================
# 13. Save Hasil Prediksi Uji
# ==============================================================================
df_out = df.iloc[idx_test].copy().reset_index(drop=True)
df_out['y_true'] = y_test
df_out['pred_prob'] = y_prob
df_out['pred_best'] = y_best

df_out.to_csv(TEST_PRED_CSV, index=False)

# ==============================================================================
# 14. Visualisasi Kurva Pembelajaran Hasil Optimasi
# ==============================================================================
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy', color='darkblue', linewidth=2)
plt.plot(history.history['val_accuracy'], label='Val Accuracy', color='crimson', linestyle='--', linewidth=2)
plt.title('Kurva Akurasi (Optimized Architecture)', fontsize=12)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, linestyle=':')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', color='darkblue', linewidth=2)
plt.plot(history.history['val_loss'], label='Val Loss', color='crimson', linestyle='--', linewidth=2)
plt.title('Kurva Loss (Optimized Architecture)', fontsize=12)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, linestyle=':')

plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, 'Learning_Curves_Optimized.png'), dpi=300)
plt.show()

print("\nOPTIMASI SELESAI — MODEL STRUKTUR TERBARU TELAH BERHASIL DIEVALUASI.")
```

```python
import os
import numpy as np
import pandas as pd
import scipy.stats as stats
from sklearn.metrics import roc_auc_score, f1_score

# ==============================================================================
# 1. LOAD DATA PENGUJIAN RIIL DARI CSV Anda
# ==============================================================================
PATH_M2 = r'model_pembanding_CNN-LSTM_1jam/test_predictions.csv'
PATH_M3 = r'model_pembanding_CNN-LSTM_1jam_optimized/test_predictions_optimized.csv'

# Validasi keberadaan file
if not os.path.exists(PATH_M2) or not os.path.exists(PATH_M3):
    raise FileNotFoundError("Salah satu atau kedua file CSV tidak ditemukan. Pastikan path sudah benar.")

df_m2 = pd.read_csv(PATH_M2)
df_m3 = pd.read_csv(PATH_M3)

# Memastikan jumlah baris data uji sama (menggunakan sampel data uji yang sama)
assert len(df_m2) == len(df_m3), "Jumlah sampel data uji antara Model 2 dan Model 3 tidak sama!"

# Ekstraksi array riil
# Catatan: Sesuaikan nama kolom jika di file 'test_predictions_optimized.csv' nama kolomnya berbeda
y_true = df_m2['y_true'].values 
prob_model2 = df_m2['pred_prob'].values
prob_model3 = df_m3['pred_prob'].values

print(f"Berhasil memuat {len(y_true)} sampel data uji.")
print(f"ROC AUC Aktual Model 2 (Eksperimen 1) : {roc_auc_score(y_true, prob_model2):.4f}")
print(f"ROC AUC Aktual Model 3 (Optimized)   : {roc_auc_score(y_true, prob_model3):.4f}\n")

# ==============================================================================
# 2. UJI SIGNIFIKANSI ROC AUC (BOOTSTRAP DELONG APPROACH)
# ==============================================================================
print("=== 1. UJI SIGNIFIKANSI PERBEDAAN ROC AUC ===")

def uji_signifikansi_auc(y_true, prob_m2, prob_m3, n_bootstraps=1000, seed=42):
    np.random.seed(seed)
    auc_diffs = []
    indices = np.arange(len(y_true))
    
    for i in range(n_bootstraps):
        # Resampling indeks dengan pengembalian (Bootstrap)
        boot_idx = np.random.choice(indices, size=len(indices), replace=True)
        
        # Penanganan darurat jika sampel bootstrap hanya berisi 1 kelas
        if len(np.unique(y_true[boot_idx])) < 2:
            continue
            
        auc_m2 = roc_auc_score(y_true[boot_idx], prob_m2[boot_idx])
        auc_m3 = roc_auc_score(y_true[boot_idx], prob_m3[boot_idx])
        auc_diffs.append(auc_m3 - auc_m2)
        
    auc_diffs = np.array(auc_diffs)
    mean_diff = np.mean(auc_diffs)
    std_diff = np.std(auc_diffs)
    
    # Menghitung Z-score dan p-value dua arah (Two-Tailed Test)
    z_stat = mean_diff / std_diff
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    
    return mean_diff, p_value

mean_diff, p_val_auc = uji_signifikansi_auc(y_true, prob_model2, prob_model3, n_bootstraps=1000)
print(f"Rata-rata Selisih AUC (Model 3 - Model 2): {mean_diff:.4f}")
print(f"Nilai p-value                             : {p_val_auc:.6f}")

if p_val_auc < 0.05:
    print("Kesimpulan: Perbedaan SIGNIFIKAN secara statistik (p < 0.05).")
    print("            Peningkatan performa Model 3 BUKAN karena faktor kebetulan.")
else:
    print("Kesimpulan: Perbedaan TIDAK signifikan secara statistik (p >= 0.05).")
    print("            Peningkatan performa kemungkinan dipengaruhi oleh noise.")
print("-" * 60)

# ==============================================================================
# 3. UJI BOOTSTRAP CONFIDENCE INTERVAL (CI 95%) UNTUK F1-SCORE
# ==============================================================================
print("\n=== 2. UJI CONFIDENCE INTERVAL (CI 95%) F1-SCORE ===")

def uji_bootstrap_f1_ci(y_true, prob_m2, prob_m3, thr_m2=0.70, thr_m3=0.80, n_bootstraps=1000, seed=42):
    np.random.seed(seed)
    f1_m2_list = []
    f1_m3_list = []
    indices = np.arange(len(y_true))
    
    for i in range(n_bootstraps):
        boot_idx = np.random.choice(indices, size=len(indices), replace=True)
        
        # Klasifikasikan biner berdasarkan threshold masing-masing program Anda
        pred_m2 = (prob_m2[boot_idx] >= thr_m2).astype(int)
        pred_m3 = (prob_m3[boot_idx] >= thr_m3).astype(int)
        
        f1_m2_list.append(f1_score(y_true[boot_idx], pred_m2, zero_division=0))
        f1_m3_list.append(f1_score(y_true[boot_idx], pred_m3, zero_division=0))
        
    # Mengambil persentil 2.5% dan 97.5% untuk menentukan rentang kepercayaan 95%
    ci_m2 = np.percentile(f1_m2_list, [2.5, 97.5])
    ci_m3 = np.percentile(f1_m3_list, [2.5, 97.5])
    
    return ci_m2, ci_m3

ci_m2, ci_m3 = uji_bootstrap_f1_ci(y_true, prob_model2, prob_model3, thr_m2=0.70, thr_m3=0.80)
print(f"95% CI F1-Score Model 2 (Threshold 0.70): [{ci_m2[0]:.4f} ke {ci_m2[1]:.4f}]")
print(f"95% CI F1-Score Model 3 (Threshold 0.80): [{ci_m3[0]:.4f} ke {ci_m3[1]:.4f}]")

# Cek Overlapping rentang keyakinan
if ci_m3[0] > ci_m2[1] or ci_m2[0] > ci_m3[1]:
    print("Kesimpulan: Interval kepercayaan kedua model TIDAK saling tumpang tindih (No Overlap).")
    print("            Perubahan nilai F1-score terbukti signifikan secara statistik.")
else:
    print("Kesimpulan: Interval kepercayaan kedua model saling tumpang tindih (Overlap).")
    print("            Secara statistik, performa F1-Score kedua arsitektur setara/stabil.")
print("==================================================")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 1. Setup Canvas Jurnal (Proporsi Vertikal)
fig, ax = plt.subplots(figsize=(8, 12), dpi=300)
ax.set_xlim(0, 10)
ax.set_ylim(0, 18)
ax.axis('off')

# 2. Fungsi Kotak Modern dengan Drop Shadow
def draw_modern_box(ax, x, y, width, height, title, subtitle, bg_color):
    # Shadow
    shadow = patches.FancyBboxPatch((x + 0.1, y - 0.1), width, height, boxstyle="round,pad=0.2", 
                                    facecolor='#d9d9d9', edgecolor='none', alpha=0.7)
    ax.add_patch(shadow)
    
    # Main Box
    rect = patches.FancyBboxPatch((x, y), width, height, boxstyle="round,pad=0.2", 
                                  facecolor=bg_color, edgecolor='#2c3e50', linewidth=1.2)
    ax.add_patch(rect)
    
    # Typography
    ax.text(x + width/2, y + height - 0.35, title, ha='center', va='center', 
            color='#ffffff' if bg_color != '#f8f9fa' else '#2c3e50', 
            fontsize=12, fontweight='bold', fontfamily='sans-serif')
    ax.text(x + width/2, y + 0.45, subtitle, ha='center', va='center', 
            color='#f1f2f6' if bg_color != '#f8f9fa' else '#576574', 
            fontsize=9.5, fontfamily='sans-serif', wrap=True)

# 3. Fungsi Panah Elegan
def draw_arrow(ax, x, y_start, length):
    ax.annotate('', xy=(x, y_start - length), xytext=(x, y_start),
                arrowprops=dict(facecolor='#7f8c8d', edgecolor='#7f8c8d', 
                                width=2, headwidth=8, headlength=10, shrink=0))

# 4. Palet Warna Berdasarkan Fase (Standar Jurnal: Data Prep -> Modeling -> Eval)
color_data = '#3498db'  # Biru Profesional (Data Prep)
color_model = '#27ae60' # Hijau (Modeling)
color_eval = '#e67e22'  # Oranye (Evaluation)

steps = [
    ("1. Raw Sensor Data", "Multivariate meteorological inputs", color_data),
    ("2. Linear Interpolation", "Handling missing values & data cleaning", color_data),
    ("3. Feature Engineering", "Diurnal cycle time transformation (Sin/Cos)", color_data),
    ("4. Sliding Window Sequencing", "Structuring 12 time-steps (1-hour history)", color_data),
    ("5. Time-Based Data Split", "Preserving temporal order (Train 80% : Test 20%)", color_data),
    ("6. Class Weight Calculation", "Addressing imbalanced rain/weather classes", color_data),
    ("7. Proposed Model Training", "Optimized Attention-CNN-LSTM Network", color_model),
    ("8. Evaluation & Output", "F1-Score threshold optimization & Predictions", color_eval)
]

# 5. Eksekusi Penggambaran
box_w = 7.0
box_h = 1.2
center_x = 1.5
current_y = 15.5
gap = 1.8

for i, (title, subtitle, color) in enumerate(steps):
    draw_modern_box(ax, center_x, current_y, box_w, box_h, title, subtitle, color)
    if i < len(steps) - 1:
        draw_arrow(ax, center_x + box_w/2, current_y - 0.05, gap - box_h - 0.1)
    current_y -= gap

# 6. Menambahkan Judul Gambar
plt.text(5.0, 17.5, "Data Processing and Modeling Pipeline", 
         ha='center', va='center', fontsize=14, fontweight='bold', color='#2c3e50', fontfamily='sans-serif')

# 7. Simpan Gambar Resolusi Tinggi
plt.tight_layout()
plt.savefig('Flowchart_Modern_SINTA1.png', bbox_inches='tight', dpi=300, transparent=False)
plt.show()

print("Flowchart modern berhasil disimpan sebagai 'Flowchart_Modern_SINTA1.png'")
```
